### Objective

To analyze patient health records using Exploratory Data Analysis (EDA) and identify patterns between age, blood sugar level, blood pressure, BMI, and disease status that can support healthcare decision-making.

```python
# ==========================================================
# Import Required Libraries
# ==========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use("ggplot")
sns.set_theme(style="whitegrid")

print("Libraries imported successfully.")
```

## Dataset Description

### Problem Statement

A hospital wants to analyze patient health risk based on age, blood sugar level, blood pressure, BMI, and disease status. The objective is to identify important health patterns using Exploratory Data Analysis.

### Dataset Features

- Patient_ID
- Age
- Sugar_Level
- BP
- BMI
- Disease_Status

### Target Variable

Disease_Status (Yes / No)

```python
# ==========================================================
# Create Healthcare Dataset
# ==========================================================

healthcare = {
    "Patient_ID":["P01","P02","P03","P04","P05","P06","P07","P08","P09","P10",
                  "P11","P12","P13","P14","P15","P16","P17","P18","P19","P20"],

    "Age":[45,32,55,28,60,40,52,35,48,30,62,38,50,27,58,42,46,34,53,29],

    "Sugar_Level":[145,98,180,90,200,120,165,105,155,95,210,115,170,88,190,130,150,100,175,92],

    "BP":[130,118,145,110,150,125,140,120,135,115,155,122,142,108,148,128,132,116,144,112],

    "BMI":[28,23,31,22,34,26,30,24,29,23,35,25,30,21,33,27,28,24,31,22],

    "Disease_Status":["Yes","No","Yes","No","Yes","No","Yes","No","Yes","No",
                      "Yes","No","Yes","No","Yes","No","Yes","No","Yes","No"]
}

df2 = pd.DataFrame(healthcare)

df2.head()
```

## Dataset Understanding

### Purpose of the Dataset

This dataset contains patient health information collected to analyze the relationship between age, blood sugar level, blood pressure, BMI, and disease status. The analysis helps identify health patterns and supports better medical decision-making.

```python
# ==========================================================
# Dataset Information
# ==========================================================

print("Shape of Dataset:")
print(df2.shape)

print("\nDataset Information:")
df2.info()

print("\nData Types:")
print(df2.dtypes)
```

```python
# ==========================================================
# Data Cleaning
# ==========================================================

print("Missing Values")
print(df2.isnull().sum())

print("\nDuplicate Records")
print(df2.duplicated().sum())
```

```python
# ==========================================================
# Check for Incorrect Entries
# ==========================================================

print("Unique Disease Status Values:")
print(df2["Disease_Status"].unique())
```

```python
# ==========================================================
# Descriptive Statistics
# ==========================================================

print(df2.describe())
```

### Interpretation

- The dataset contains **20 patient records** and **6 features**.
- No missing values or duplicate records were found.
- Disease_Status contains only valid categories (**Yes** and **No**).
- The dataset is clean and ready for further analysis.

## Numerical and Categorical Analysis

```python
# ==========================================================
# Numerical Analysis
# ==========================================================

print("Average Age:", df2["Age"].mean())
print("Average Sugar Level:", df2["Sugar_Level"].mean())
print("Average Blood Pressure:", df2["BP"].mean())
print("Average BMI:", df2["BMI"].mean())

print("\nDisease Status Count:")
print(df2["Disease_Status"].value_counts())
```

## Outlier Detection

```python
# ==========================================================
# Box Plot for Outlier Detection
# ==========================================================

plt.figure(figsize=(6,4))
plt.boxplot(df2["Sugar_Level"])
plt.title("Box Plot of Sugar Level")
plt.ylabel("Sugar Level")
plt.show()
```

## Data Visualization

```python
# ==========================================================
# Histogram
# ==========================================================

plt.figure(figsize=(6,4))
plt.hist(df2["Age"], bins=6, edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.show()
```

```python
# ==========================================================
# Bar Chart
# ==========================================================

plt.figure(figsize=(6,4))
df2["Disease_Status"].value_counts().plot(kind="bar", color=["green","red"])

plt.title("Disease Status Distribution")
plt.xlabel("Disease Status")
plt.ylabel("Number of Patients")
plt.show()
```

```python
# ==========================================================
# Scatter Plot
# ==========================================================

plt.figure(figsize=(6,4))

plt.scatter(df2["Sugar_Level"], df2["BMI"])

plt.title("Sugar Level vs BMI")
plt.xlabel("Sugar Level")
plt.ylabel("BMI")

plt.show()
```

```python
# ==========================================================
# Pie Chart
# ==========================================================

plt.figure(figsize=(5,5))

df2["Disease_Status"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Disease Status Percentage")
plt.ylabel("")
plt.show()
```

```python
# ==========================================================
# Heatmap
# ==========================================================

plt.figure(figsize=(6,4))

sns.heatmap(
    df2.select_dtypes(include=np.number).corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

## Observations

1. Patients with higher blood sugar levels generally have higher BMI values.
2. Older patients are more likely to have higher blood pressure.
3. The dataset contains an equal number of patients with and without disease.
4. No missing values or duplicate records were found.
5. The heatmap indicates positive relationships among age, sugar level, blood pressure, and BMI.

## Conclusion

The Exploratory Data Analysis shows that higher blood sugar, blood pressure, and BMI are associated with patients diagnosed with the disease. The dataset is clean and suitable for analysis. These findings can help hospitals identify high-risk patients and support preventive healthcare decisions.
