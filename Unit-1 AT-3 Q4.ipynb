## Assessment Tool 3
### Question 4 – Banking Loan Approval Dataset

### Objective

To perform Exploratory Data Analysis (EDA) on a banking loan approval dataset to understand how income, credit score, loan amount, and employment type influence loan approval decisions.

```python
# ==========================================================
# Import Required Libraries
# ==========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use("ggplot")
sns.set_theme(style="whitegrid")

print("Libraries imported successfully.")
```

## Dataset Description

### Problem Statement

A bank wants to analyze loan approval decisions based on income, credit score, loan amount, employment type, and approval status.

### Dataset Features

- Customer_ID
- Income
- Credit_Score
- Loan_Amount
- Employment_Type
- Loan_Status

### Target Variable

Loan_Status (Approved / Rejected)

```python
# ==========================================================
# Create Banking Loan Dataset
# ==========================================================

bank = {
    "Customer_ID":["C01","C02","C03","C04","C05","C06","C07","C08","C09","C10",
                   "C11","C12","C13","C14","C15","C16","C17","C18","C19","C20"],

    "Income":[45000,25000,60000,22000,50000,30000,70000,28000,55000,24000,
              65000,35000,48000,26000,75000,32000,52000,27000,68000,23000],

    "Credit_Score":[720,580,750,560,710,600,780,590,730,570,
                    760,620,700,585,790,610,725,575,770,555],

    "Loan_Amount":[200000,150000,300000,120000,250000,180000,350000,160000,270000,130000,
                   320000,200000,240000,140000,400000,190000,260000,150000,330000,125000],

    "Employment_Type":["Salaried","Self-employed","Salaried","Unemployed","Salaried",
                       "Self-employed","Salaried","Self-employed","Salaried","Unemployed",
                       "Salaried","Self-employed","Salaried","Unemployed","Salaried",
                       "Self-employed","Salaried","Self-employed","Salaried","Unemployed"],

    "Loan_Status":["Approved","Rejected","Approved","Rejected","Approved","Rejected",
                   "Approved","Rejected","Approved","Rejected","Approved","Rejected",
                   "Approved","Rejected","Approved","Rejected","Approved","Rejected",
                   "Approved","Rejected"]
}

df4 = pd.DataFrame(bank)

df4.head()
```

```python
print("Shape:", df4.shape)

print("\nDataset Information")
df4.info()

print("\nData Types")
print(df4.dtypes)
```

```python
print("Missing Values")
print(df4.isnull().sum())

print("\nDuplicate Records")
print(df4.duplicated().sum())

print("\nEmployment Types")
print(df4["Employment_Type"].unique())
```

```python
df4.describe()
```

```python
print("Average Income:", df4["Income"].mean())
print("Average Credit Score:", df4["Credit_Score"].mean())
print("Average Loan Amount:", df4["Loan_Amount"].mean())

print("\nLoan Status")
print(df4["Loan_Status"].value_counts())
```

```python
plt.figure(figsize=(6,4))
df4["Loan_Status"].value_counts().plot(kind="bar")
plt.title("Loan Approval Status")
plt.xlabel("Status")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df4["Income"], bins=6, edgecolor="black")
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df4["Credit_Score"], df4["Loan_Amount"])
plt.title("Credit Score vs Loan Amount")
plt.xlabel("Credit Score")
plt.ylabel("Loan Amount")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(df4["Credit_Score"])
plt.title("Credit Score Box Plot")
plt.show()
```

```python
plt.figure(figsize=(5,5))
df4["Employment_Type"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Employment Type Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df4.select_dtypes(include=np.number).corr(),
            annot=True,
            cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

## Observations

1. Customers with higher credit scores are generally approved for larger loans.
2. Salaried customers have a higher approval rate in this dataset.
3. No missing values or duplicate records were found.
4. Income and loan amount show a positive relationship.
5. Credit score is an important factor in loan approval.

## Conclusion

The EDA indicates that income, credit score, and employment type play an important role in loan approval decisions. These insights can help banks improve loan evaluation and reduce financial risk.
